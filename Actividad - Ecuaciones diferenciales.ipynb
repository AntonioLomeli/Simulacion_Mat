1. Resolver numéricamente con `solve_ivp` la siguiente ecuación diferencial:
$$\frac{dx}{dt}=\frac{x}{t+1}$$

```python
import numpy as np
from scipy.integrate import solve_ivp, odeint
import matplotlib.pyplot as plt
```

```python
def fun(x,t):
    return x/(t+1)

tspan=[0,50]
x0= np.array([2])

x=solve_ivp(fun, tspan, x0, max_step=0.05)
```

Grafique con condición inicial $x_0 = 2$

```python
plt.figure()
plt.plot(x.t, x.y[0,:])
```

2. Encuentre la solución numérica para un oscilador armónico simple:

$$
\frac{d^2 x}{dt^2} +  \omega_{0}^2 x = 0.
$$
Donde $x_0=0.5$, $\dot{x}_0=0.1$ y $\omega_0=3$.

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
solve_ivp#? Solo puede resolver ecuaciones diferenciales de primer orden, esta es de segundo orden, hay que encontrar la primera
```

```python
def dX(t,X):
    return np.array([X[1],-(3**2)*X[0]])

tspan=[0,50] # Desde donde empieza y hasta donde termina
X= np.array([0.5,0.1]) #Velocidad y posicion iniciales

x2=solve_ivp(dX, tspan, X)
```

Grafique

```python
plt.figure()
plt.plot(x2.t, x2.y[0,:])
```

Asigne a los parámetros de la solución analítica $C_1=x_0$ y $C_2=\frac{\dot{x}_0}{\omega_0}$.

```python
def posicion (t,w0=3,x0=0.5,v0=0.1):
    return (x0*np.cos(w0*t)) + ((v0/w0)*np.sin(w0*t))
```

```python
t=np.linspace(0,50,500)

plt.figure()
plt.plot(t,posicion(t),c='g')
```

Redefina la función para graficarla (con la palabra reservada ***def***) en un intervalo de tiempo de 0 a 10 segundos. Comparela en un mismo grafico con la solución numérica obtenida.

```python
def dX(t,X):
    return np.array([X[1],-(3**2)*X[0]])

tspan=[0,10] # Desde donde empieza y hasta donde termina
X= np.array([0.5,0.1]) #Velocidad y posicion iniciales

x4=solve_ivp(dX, tspan, X)
```

```python
t=np.linspace(0,10,1000)

plt.figure()
plt.plot(t,posicion(t),c='g', label='Analítica', lw=3)
plt.plot(x4.t, x4.y[0,:], c='b', label='Numérica')
plt.grid()
plt.legend(loc='best')
```
