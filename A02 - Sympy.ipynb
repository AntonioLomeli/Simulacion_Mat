**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

def fun(x):
    return 1260+(117*x)-(9*x**2)

sp.var('x')
dif=sp.diff(fun(x),x)
pc=sp.solve(dif,x)
pc2=float(pc[0])
y=float(fun(pc[0]))

xn=np.linspace(-23.5,36.5,200)
yn=fun(xn)

ddf=sp.diff(dif,x)
prueba=ddf.subs(x,pc[0])

if prueba==0:
    print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y no es ni maximo ni mínimo')
elif prueba>0:
    if (y<fun(pc2-2)) and (y<fun(pc2+2)) and (y<fun(pc2-150)) and (y<fun(pc2+180)):
        print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un mínimo global')
    else:
        print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un mínimo local')
elif prueba<0:
    if (y>fun(pc2-2)) and (y>fun(pc2+2)) and (y>fun(pc2-150)) and (y>fun(pc2+180)):
        print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un máximo global')
    else:
        print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un máximo local')
    
plt.figure()
plt.plot(xn,yn,'r')
plt.grid()

plt.plot(pc[0],y,'*g',ms=15)
```

$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

def fun(x):
    return 6-(2*x)+((18-(2*(x**2)))**(1/2))

sp.var('x')
dif=sp.diff(fun(x),x)
pc=sp.solve(dif,x)

xn=np.linspace(-3,3,200)
yn=fun(xn)
plt.figure()
plt.plot(xn,yn,'g')
plt.grid()

for i in pc:
    if i>=-3 and i<=3:
        pc2=round(i,3)
        y=round(fun(pc2),3)
        plt.plot(pc2,y,'*r',ms=10)

        ddf=sp.diff(dif,x)
        prueba=ddf.subs(x,pc2)

        if prueba==0:
            print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y no es ni maximo ni mínimo')
        elif prueba>0:
            if (y<fun(pc2-0.01)) and (y<fun(pc2+0.01)) and (y<fun(pc2+5)):
                print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un mínimo global')
            else:
                print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un mínimo local')
        elif prueba<0:
            if (y>fun(pc2-0.5)) and (y>fun(pc2+0.5)) and (y>fun(pc2+5)):
                print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un máximo global')
            else:
                print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y es un máximo local')
```

El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

***
#### Aquí empiza la solución
Si dejamos la base como variable independiente x y los lados en función de lo que mide la base tendremos dos ecuaciones que parten de una ecuación original, la ecuación original es:

$$10=x+2y$$ 
donde x es la base y y la medida de los lados

La ec que deja los dos lados iguales en función de la base es:

$$ y=5-(x/2) $$

Esto nos va a servir para calcular la altura del cono, si partimos la base a la mitad formamos un triángulo rectángulo, la altura sería un cateto, entonces usando el teorema de pitágoras:

$$ h=sqrt(hipotenusa^2-cateto^2) $$

La hipotenusa siempre va a ser el lado del triángulo rectángulo, porque por el ***teorema de la desigualdad del triángulo*** [Portal Academico CCH UNAM](https://portalacademico.cch.unam.mx/matematicas2/geometria-del-triangulo/desigualdad-triangulo)  sabemos que dos lados cualesquiera de un triángulo deben medir más que el otro lado restante, entonces, si los lados iguales miden 3 y 3 cm, su suma es mayor a la longitud de la base (4 cm). Sin embargo, si la base mide 6 cm, la longitud de los lados es de 2cm cada uno y 2+2 no es mayor a 6. Entonces, el dominio de "x" (longitud de la base del triángulo será):

$$ 0<=x<5 $$

```python
#Declaramos las variables y encontramos la ec de los lados con respecto a la longitud de la base (x)
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

sp.var('x')
sp.var('y')

ec_general=x+(2*y)-10
ec_lados=sp.solve(ec_general,y)
ec_lados
```

```python
#En este paso resolvemos el teorema de pitágoras, donde el lado es la hipotenusa
base=x/2
lado=ec_lados[0]

altura=sp.simplify((lado**2)-(base**2))
altura=(altura**(1/2))
altura
```

Ahora que tenemos la altura queda dejar el radio de la base del cono en función de la medida de la base del triángulo, es decir, de su circunferencia:

$$ circun=pi*r $$
$$ r=circun/pi $$

Con eso, podemos tener la función V(x) o volumen de un cono en función de la base del triángulo que lo origina.

```python
# Declaramos que r es igual a circun/pi

r=(x/(sp.pi))

# Formamos la función V(x)

V=sp.simplify((1/3)*((sp.pi)*(r**2)*(altura)))
print(V)
print('\n')

#Obtenemos la derivada de V y sus soluciones para empezar el proceso de optimización, además graficamos

dV=sp.diff(V,x)
pc=sp.solve(dV,x)

Vn=sp.lambdify(x,V,'numpy')
xn=np.linspace(0,5,200)
yn=Vn(xn)

plt.figure()
plt.plot(xn,yn,'r')
plt.grid()

# Usando los puntos críticos obtenemos el volumen que con esa base se forma y graficamos los puntos

for i in pc:
    volumen=float(V.subs(x,i))
    volumen=round(volumen,2)
    mLados=int(lado.subs(x,i)) #Cuanto mide el lado con la medida de la base
    ddV=sp.diff(dV,x)
    prueba=ddV.subs(x,i)
    i=int(i)
    
    plt.plot(i,volumen,'^b',ms=12)
    
    if prueba==0:
        print(f"El punto crítico que está en ( {i} , {volumen} ) no es ni máximo ni mínimo")
    elif prueba>0:
        print(f'El volumen mínimo de {volumen}cm3 se forma cuando:\nLa base del triángulo mide {i}cm y sus lados {mLados}cm\n')
    elif prueba<0:
        print(f'El volumen máximo de {volumen}cm3 se forma cuando:\nLa base del triángulo mide {i}cm y sus lados {mLados}cm')
    
```

La respuesta es, que el ***volumen máximo*** del cono que se formaría al rotar un triángulo isósceles de 10 cm de perímetro sobre su base ***es de 3.8 cm3.***
Y su ***base debe medir 4 cm*** y sus dos ***lados iguales 3cm.***

Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño?

Como una portería es un rectángulo, tomaré la "base" (la parte de arriba) como variable independiente y los lados que la sostienen quedarán en función de la longitud de la parte de arriba. Las ecuación original es:

$$ 6=x+2y $$

"x" representa la base y "y" la altura. Si despejamos "y":

$$ y=3-(x/2)$$

La ecuación del área de un rectángulo es:
$$A=b*h$$

Si sustituimos los valores de "b" con "x" y los de "h" con "y" podemos formar la función A(x):
   
$$ A(x)=x*y $$
$$ A(x)=x*(3-(x/2)) $$
$$ A(x)=3x-(x^2/2) $$

Teniendo la función $ A(x)=3x-(x^2/2) $ podemos hacer un proceso de optimización con la primera y segunda derivada de la función A(x).


```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

def rect(x):
    return (3*x)-((x**2)/2)

sp.var('x')
dif=sp.diff(rect(x),x)
pc=sp.solve(dif,x)

xn=np.linspace(0,6,200)
yn=rect(xn)
plt.figure()
plt.plot(xn,yn,'y')
plt.grid()

for i in pc:
    if i>=0 and i<=6:
        pc2=round(i,2)
        y=round(rect(pc2),2)
        plt.plot(pc2,y,'*g',ms=12)
        altura=round(3-pc2/2,2)

        ddf=sp.diff(dif,x)
        prueba=ddf.subs(x,pc2)

        if prueba==0:
            print(f'El punto crítico se encuentra en ( {pc2} , {y} ) y no es ni maximo ni mínimo')
        elif prueba>0:
            print(f'Con una base de {pc2}m y altura de {altura}m\nTendríamos el ÁREA MÍNIMA que es de {y} metros')
        elif prueba<0:
            print(f'Con una base de {pc2}m y altura de {altura}m\nTendríamos el ÁREA MÁXIMA que es de {y} metros')
```
